```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
# there are many entry in total_sqft column that are in range and datatype of column is object
df[~df['total_sqft'].apply(is_float)]
```

```python
# taking average of range in total_sqft column

def convert_sqft_to_num(x):
    ranges = x.split('-')
    if len(ranges) == 2:
        return (float(ranges[0])+float(ranges[1]))/2
    try:
        return float(x)
    except:
        return None  
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)
```

```python
df = df[df['total_sqft'].notnull()]
```

### Adding new feature called price per square feet

```python
df['price_per_sqft'] = (df['price']*100000)/df['total_sqft']
df
```

## Applying dimensionality reduction to reduce the number of locations

```python
df['location'] = df['location'].apply(lambda x: str(x).strip() if isinstance(x, str) else x)

location_value_count = df['location'].value_counts()
location_value_count
```

```python
df['location'].nunique()
```

### Any location having value count of less than 10 should be tagged as 'other' location

```python
location_value_count_less_than_10 = location_value_count[location_value_count<=10]
df['location'] = df['location'].apply(lambda x: 'other' if x in location_value_count_less_than_10 else x)
df['location'].nunique()
```

### Normally sqft per bedroom is 300, if 2bhk apartment has 400 sqft then that can removed as outlier

```python
df[df['total_sqft']/df['bhk'] < 300]
```

```python
df = df[~(df['total_sqft']/df['bhk'] < 300)]
```

```python
df = df.drop(['size','price_per_sqft'],axis='columns')
```

### One hot encoding for location and area_type

```python
# dummies for location
dummies1 = pd.get_dummies(df['location']).astype(int)
```

```python
df = pd.concat([df,dummies1.drop('other', axis='columns'),], axis='columns')
df
```

```python
df = df.drop(['location'], axis='columns')
```

### Model selection using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

def find_best_model_using_gridsearchcv(x, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

find_best_model_using_gridsearchcv(x, y)
```
